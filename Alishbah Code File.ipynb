```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, precision_score, recall_score, roc_auc_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier

```

```python
df= pd.read_csv("breast-cancer.csv")
df.head()
```

```python
df.value_counts("diagnosis")
```

```python
df.drop('id', axis=1, inplace=True)
df.shape
```

```python
# 2. Data Preprocessing
# Encoding categorical variables
le = LabelEncoder()
df['diagnosis'] = le.fit_transform(df['diagnosis'])

# Scaling numerical features
scaler = StandardScaler()
df[df.columns[1:]] = scaler.fit_transform(df[df.columns[1:]])
```

```python
# 3. Feature Selection
selector = SelectKBest(score_func=f_classif, k=10)
selector.fit(df[df.columns[1:]], df['diagnosis'])
selected_features = df.columns[1:][selector.get_support()]
```

```python
df.head(8)
```

```python
# 4. Model Training
X_train, X_test, y_train, y_test = train_test_split(df[selected_features], df['diagnosis'], test_size=0.2, random_state=42)
```

```python
# Logistic Regression
lr = LogisticRegression()
lr.fit(X_train, y_train)
```

```python
# Decision Tree
dt = DecisionTreeClassifier()
dt.fit(X_train, y_train)
```

```python
# Random Forest
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
```

```python
# 5. Model Evaluation
models = [lr, dt, rf]
for model in models:
    y_pred = model.predict(X_test)
    print(f"Model: {model.__class__.__name__}")
    print(f"Accuracy: {accuracy_score(y_test, y_pred)}")
    print(f"Precision: {precision_score(y_test, y_pred)}")
    print(f"Recall: {recall_score(y_test, y_pred)}")
    print(f"ROC AUC: {roc_auc_score(y_test, y_pred)}")
    print("------------------------")
```

```python
import pickle

# Save Logistic Regression Model
with open('logistic_regression_model.pkl', 'wb') as f:
    pickle.dump(lr, f)

# Save Decision Tree Model
with open('decision_tree_model.pkl', 'wb') as f:
    pickle.dump(dt, f)

# Save Random Forest Model
with open('random_forest_model.pkl', 'wb') as f:
    pickle.dump(rf, f)
```
